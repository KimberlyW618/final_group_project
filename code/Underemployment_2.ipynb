```python
import pandas as pd
```

```python
#Load the data
df = pd.read_csv('raw_data/Underemployment.csv')
df.head(10)
```

```python
df.dtypes
```

```python
#Change the names
df = df.rename(columns={"hign_school":"high_school"})
```

```python
df
```

```python
cols = ["all", "women_b_degree", "men_b_degree", "less_than_hs", "high_school", "some_college", "bachelor_degree", "advanced_degree", "women", "men"]
df[cols] = df[cols].apply(lambda x: round(x * 100, 2))
df
```

```python
#Output clean data
df.to_csv('new_data/Underemployment.csv',index=False)
```
