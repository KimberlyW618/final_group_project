```python
#Import dependencies
import pandas as pd
from datetime import datetime
```

```python
#Load the data
df = pd.read_csv('../raw_data/Labor-Force_participation.csv')
df.head(10)
```

```python
#Check the data types
df.dtypes
```

```python
#Filter the year
df.loc[df["Year"]>1987]
```

```python
#Drop the NA
df_labor_force_participation=df.dropna()
df_labor_force_participation.head(10)
```

```python
#Select columns
df_labor_force_participation=df_labor_force_participation[["Year","All","Less than HS","High school","Some college","Bachelor's degree","Advanced degree","Women","Men"]] 
```

```python
df_labor_force_participation.head()
```

```python
df_labor_force_participation.tail(10)
```

```python
#Change the names
df_labor_force_participation=df_labor_force_participation.rename(columns={"Year":"year","All":"all","Less than HS":"less_than_hs","High school":"hign_school","Some college":"some_college","Bachelor's degree":"bachelor_degree","Advanced degree":"advanced_degree","Women":"women","Men":"men"})
```

```python
df_labor_force_participation
```

```python
#Group by years and get the average number 
df_labor_force_participation=round(df_labor_force_participation.groupby("year").mean().astype(float),3)
```

```python
#Value sort by year
df_labor_force_participation.sort_values("year",ascending=False)
```

```python
#Output clean data
df_labor_force_participation.to_csv('../clean_data/Labor-Force_Participation.csv',index=True)
```
