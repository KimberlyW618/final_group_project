```python
import pandas as pd
```

```python
#Load the data
df = pd.read_csv('../raw_data/unemployment.csv')
df.head(10)
```

```python
#Check the data types
df.dtypes
```

```python
#Filter the year
df=df.loc[(df["Year"]>1987)&(df["Year"]<2021)]
```

```python
#Drop the NA
df_umemployment=df.dropna()
df_umemployment.head(10)
```

```python
#Select columns
df_umemployment=df_umemployment[["Year","All","Women Bachelor's degree","Men Bachelor's degree","Less than HS","High school","Some college","Bachelor's degree","Advanced degree","Women","Men"]] 
```

```python
df_umemployment.head()
```

```python
df_umemployment.tail()
```

```python
#Change the names
df_umemployment=df_umemployment.rename(columns={"Year":"year","All":"all","Women Bachelor's degree":"women_b_degree","Men Bachelor's degree":"men_b_degree","Less than HS":"less_than_hs","High school":"hign_school","Some college":"some_college","Bachelor's degree":"bachelor_degree","Advanced degree":"advanced_degree","Women":"women","Men":"men"})
```

```python
df_umemployment
```

```python
#Group by years and get the average number 
df_umemployment=round(df_umemployment.groupby("year").mean().astype(float),3)
```

```python
#Value sort by year
df_umemployment.sort_values("year",ascending=False)
```

```python
#Output clean data
df_umemployment.to_csv('../clean_data/unemployment_clean.csv',index=True)
```
