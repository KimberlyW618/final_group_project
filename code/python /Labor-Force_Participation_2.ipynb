```python
#Import dependencies
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
#Load the data
df = pd.read_csv('raw_data/Labor-Force_participation.csv')
df.head(10)
```

```python
#Check the data types
df.dtypes
```

```python
#Change the names
df = df.rename(columns={"hign_school":"high_school"})
```

```python
df.head()
```

```python
cols = ['all', 'less_than_hs', 'high_school', 'some_college', 'bachelor_degree', 'advanced_degree', 'women', 'men']
df[cols] = df[cols].apply(lambda x: round(x * 100, 2))
df
```

```python
#Output clean data
df.to_csv('new_data/Labor-Force_Participation.csv',index=False)
```
