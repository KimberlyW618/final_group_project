```python
import pandas as pd
```

```python
#Load the data
df = pd.read_csv('../raw_data/Wages_By_Ed.csv')
df.head(10)
```

```python
#Check the data types
df.dtypes
```

```python
#Filter the year
df=df.loc[(df["Date"]>1986)&(df["Date"]<2021)]
```

```python
#Select columns
df_wages_by_education=df[["Date","Less than HS","High school","Some college","Bachelor's degree","Advanced degree","Less than HS Share","High school Share","Some college Share","Bachelor's degree Share","Advanced degree Share"]] 
```

```python
df_wages_by_education.head()
```

```python
df_wages_by_education.tail()
```

```python
#Change the names
df_wages_by_education=df_wages_by_education.rename(columns={"Date":"year","Less than HS":"less_than_hs","High school":"hign_school","Some college":"some_college","Bachelor's degree":"bachelor_degree","Advanced degree":"advanced_degree","Less than HS Share":"less_than_hs_share","High school Share":"hign_school_share","Some college Share":"some_college_share","Bachelor's degree Share":"bachelor_degree_share","Advanced degree Share":"advanced_degree_share"})
```

```python
df_wages_by_education
```

```python
#Group by years and get the average number 
df_wages_by_education=round(df_wages_by_education.astype(float),3)
```

```python
df_wages_by_education.head()
```

```python
df_wages_by_education.dtypes
```

```python
df_wages_by_education["year"]
```

```python
wages_by_education = df_wages_by_education.astype({"year": "int"})
wages_by_education.head()
```

```python
wages_by_education.dtypes
```

```python
#Output clean data
wages_by_education.to_csv('../clean_data/Wages_By_Ed.csv',index=False)
```
