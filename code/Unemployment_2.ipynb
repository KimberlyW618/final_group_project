```python
import pandas as pd
import numpy as np
import io
```

```python
#Load the data
df = pd.read_csv('raw_data/Unemployment.csv')
df.head(10)
```

```python
df = df.rename(columns={"hign_school":"high_school"})
```

```python
#Check the data types
df.dtypes
```

```python
#Value sort by year
dfs = df.sort_values("year",ascending=False)
```

```python
cols = ["all", "women_b_degree", "men_b_degree", "less_than_hs", "high_school", "some_college", "bachelor_degree", "advanced_degree", "women", "men"]
```

```python

df[cols] = df[cols].apply(lambda x: round(x * 100, 2))
```

```python
df.to_csv('new_data/Unemployment.csv', index = False)
```
